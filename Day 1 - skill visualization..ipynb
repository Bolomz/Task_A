```python
#Import all used modules.
import pandas as pd
from pandas import Series, DataFrame

import numpy as np

import matplotlib.pyplot as plt
```

```python
Skills = pd.read_csv('Skill_Sets.csv')
skills = Skills.drop(['Unnamed: 0'], axis=1)
skills
```

```python
skills_=skills.groupby('Name', as_index=False).count()
skills_
```

```python
skills_.columns = ['Name','Number of skills']
skills_
```

```python
#Changing the datatype in order to read their individual values.
Number = [tuple(x) for x in skills_[['Name']].values]
z = np.arange(len(Number))
Freq = list(skills_['Number of skills'].values)

#Plotting bar chart of Frequency against Skill Set.
plt.bar(z, [int(x) for x in Freq], align='center', alpha=0.25, color='gold')
plt.xticks(z, Number, rotation=90, fontsize=16)
plt.yticks(fontsize=7)
plt.xlabel('\nName', color='b', fontsize=14)
plt.ylabel('Number of skills\n', color='b', fontsize=14)
plt.title('\n\nNumber of Skills per person\n\n', fontsize=20, color='r')

plt.show()
```

```python
#From the plot, it can be deduced that Bisola has the most skills with 8 skills,
#Followed by Oluwaferanmi and Aishat with 6 skills each,
#Abdulbaqqi with 4 skills,
#and lastly Theophilus with 3 skills.
```

```python
skill = skills.groupby('Skill Set', as_index=False).count()
skill
```

```python
Freq_ = skill[['Skill Set', 'Name']]
Freq_.columns= ['Skill Set', 'Frequency']
Freq_
```

```python
#Changing the datatype in order to read their individual values.
Skill = [tuple(x) for x in Freq_[['Skill Set']].values]
z = np.arange(len(Skill))
Freq = list(Freq_['Frequency'].values)

#Plotting bar chart of Frequency against Skill Set.
plt.bar(z, [int(x) for x in Freq], align='center', alpha=0.5, color='b')
plt.xticks(z, Skill, rotation=90, fontsize=12)
plt.yticks(fontsize=7)
plt.xlabel('\nSkill', color='r')
plt.ylabel('Frequency\n', color='r')
plt.title('Frequency of Skills\n\n', fontsize=14, color='r')

plt.show()
```

```python
#The plot shows that python is the most learned skill among the data sets with 5 people
#C, CSS,HTML, Java, Javascript, Machine learning, Microsoft word and excel are next with 2 people each
#And the remaining skills are unique to the individuals.
```
