```python
#Importing modules to be used in the visualization

import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt

import numpy as np
```

```python
#Read .csv file
read = pd.read_csv('CAvideos.csv')


#Displaying .csv file
read
```

```python
#Viewing our .csv file in its tabular form

CA = DataFrame(read)
```

```python
#Displaying the index, title, views, likes, dislikes and comment_count.
Info = CA[['video_id','title','views', 'likes', 'dislikes', 'comment_count']]
```

```python
#Display index rows and title and views columns
Views = CA[['title', 'views']]

#Display the Views in descending order.
V = Views.sort_index(by=['views'], ascending=False)

#Display only the first 10 indexes in descending order.
Views_ = V[:10]
```

```python
#Changing the datatype in order to read their individual values.
Title = [tuple(x) for x in Views_[['title']].values]
y = np.arange(len(Title))
No_of_views = list(Views_['views'].values)

#Plotting bar chart of Views against title
plt.bar(y, [int(x) for x in No_of_views], align='center', alpha=0.5, color='b')
plt.xticks(y, Title, rotation=90, fontsize=7)
plt.yticks(fontsize=7)
plt.xlabel('\nTitle', color='g', fontsize=11)
plt.ylabel('Views\n', color='g')
plt.title('Most viewed videos\n\n', fontsize=14, color='g')

plt.show()
```

```python
#Display index rows and title and likes columns
Likes = CA[['title', 'likes']]

#Display in descending order.
L = Likes.sort_index(by=['likes'], ascending=False)

#Display only the first 10 indexes in descending order.
Likes_ = L[:10]
```

```python
#Changing the datatype in order to read their individual values.
Title = [tuple(x) for x in Views_[['title']].values]
z = np.arange(len(Title))
No_of_likes = list(Likes_['likes'].values)

#Plotting bar chart of Likes against title
plt.bar(z, [int(x) for x in No_of_likes], align='center', alpha=0.25, color='g')
plt.xticks(z, Title, rotation=90, fontsize=7)
plt.yticks(fontsize=7)
plt.xlabel('\nTitle', color='r')
plt.ylabel('Likes\n', color='r')
plt.title('Top liked videos\n\n', fontsize=14, color='r')

plt.show()
```

```python
#Display index rows and title and comment_count columns
Comments = CA[['title', 'comment_count']]

#Display in descending order.
C = Comments.sort_index(by=['comment_count'], ascending=False)

#Display only the first 10 indexes in descending order.
Comments_ = C[:10]
```

```python
#Changing the datatype in order to read their individual values.
Title = [tuple(x) for x in Comments_[['title']].values]
c = np.arange(len(Title))
No_of_comments = list(Comments_ ['comment_count'].values)

#Plotting bar chart of Comments against title
plt.bar(c, [int(x) for x in No_of_comments], align='center', alpha=0.5, color='purple')
plt.xticks(c, Title, rotation=90, fontsize=7)
plt.yticks(fontsize=7)
plt.xlabel('\nTitle', color='b')
plt.ylabel('Comments\n', color='b')
plt.title('Most commented videos\n\n', fontsize=14, color='b')

plt.show()
```

```python
#Display index rows and title and dislikes columns
Dislikes = CA[['title', 'dislikes']]

#Display in descending order.
D = Dislikes.sort_index(by=['dislikes'], ascending=False)

#Display only the first 10 indexes in descending order.
Dislikes_ = D[:10]
```

```python
#Changing the datatype in order to read their individual values.
Title = [tuple(x) for x in Dislikes_[['title']].values]
d = np.arange(len(Title))
No_of_dislikes = list(Dislikes_ ['dislikes'].values)

#Plotting bar chart of Dislikes against title
plt.bar(d, [int(x) for x in No_of_dislikes], align='center', alpha=1, color='r')
plt.xticks(d, Title, rotation=90, fontsize=7)
plt.yticks(fontsize=7)
plt.xlabel('\ntitle', color='b')
plt.ylabel('Dislikes\n', color='b')
plt.title('Most disliked videos\n\n', fontsize=14, color='b')

plt.show()
```

```python
#Display index rows and title, likes and views columns
Top1 = CA[['title', 'likes', 'views']]

#Display in descending order.
T = Top1.sort_index(by=['likes','views'], ascending=False)

#Display only the first 10 indexes in descending order.
T_=T[:10]

T_
```

```python
#Saving to separate csv files.
Views_.to_csv('top_10_views.csv', encoding = 'utf-8')
Likes_.to_csv('top_10_likes.csv', encoding = 'utf-8')
Dislikes_.to_csv('top_10_dislikes.csv', encoding = 'utf-8')
Comments_.to_csv('top_10_commented.csv', encoding = 'utf-8')
T_.to_csv('Billboard_winners.csv', encoding ='utf-8')
```
